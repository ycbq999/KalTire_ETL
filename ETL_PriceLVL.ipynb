```python
#set current working directory

import os
os.chdir('C:\\Users\\sochen\\OneDrive - Kal Tire\\Documents\\PowerBI report\\Pricing Record Power BI')


DATA_PATH = 'datasource\\output_files\\all_MPR_trimmed2.csv' 
WORKING_DATA_PATH = 'Python_Code_ETL\\data\\Price LVL.csv'

import pandas as pd
import numpy as np

```

```python
DF1 = pd.read_csv(DATA_PATH)
DF1.head()
```

```python
# DF1.head()
df1 = DF1[['InternalCode','EDSNC','Tread','System_Date']]
df1['SKU #'] = df1['InternalCode'].str.lstrip('0')

df1.head()
```

```python
# df2 = df1[df1['MFGCode']=='4230']
# df2.head()

df2 = pd.read_csv(WORKING_DATA_PATH)
```

```python
df1['System_Date'] = pd.to_datetime(df1['System_Date'])

df1['SKU #'] = df1['SKU #'].astype(str)
df2['SKU #'] = df2['SKU #'].astype(str)
```

```python
import warnings
warnings.filterwarnings("ignore")

deadline_date = ['2019-01-31','2020-01-31','2021-01-31','2022-01-31','2022-07-31','2023-01-31','2022-09-30']

for date in deadline_date:
    target_date = pd.to_datetime(date)

    for index, row in df2.iterrows():
        sku = row['SKU #']
        #print(sku)
        
        df_sku = df1[df1['SKU #'] == sku]

        # print(df_sku)
        if not df_sku.empty:
            df_sku['System_Date'] = pd.to_datetime(df_sku['System_Date'])
            df_sku = df_sku[df_sku['System_Date'] <= target_date]

            if not df_sku.empty:
                df_sku['date_diff'] = abs(df_sku['System_Date'] - target_date)
                closest_date_row = df_sku.sort_values('date_diff').iloc[0]
            
                df2.loc[index, 'EDSNC '+date] = closest_date_row['EDSNC']
        # print(df2.loc[index, 'EDSNC 2019-01'])
```

```python
# export to csv
df2.to_csv('Python_Code_ETL\\Result\\PriceLVL\\Price LVL.csv', index=False)
```
